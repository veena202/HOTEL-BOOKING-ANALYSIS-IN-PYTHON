# data analysis

```python
cancelled_per=data['is_canceled'].value_counts(normalize= True)
print(cancelled_per)
plt.figure(figsize=(4,5))
plt.title('Reservation status count')
plt.bar(['not canceled','canceled'],data['is_canceled'].value_counts())
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel', hue ='is_canceled', data= data)
plt.show()
```

```python
Resort_hotel = data[data['hotel'] == 'Resort Hotel']
print(resort_hotel['is_canceled'].value_counts(normalize=True))
```

```python
ciity_hotel = data[data['hotel'] == 'City Hotel']
print(ciity_hotel['is_canceled'].value_counts(normalize=True))
```

```python
Resort_hotel= Resort_hotel.groupby('reservation_status_date')[['adr']].mean()
City_hotel= Resort_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Aggregating ADR over 'reservation_status_date'
Resort_hotel_avg =Resort_hotel.groupby('reservation_status_date')['adr'].mean()
city_hotel_avg = ciity_hotel.groupby('reservation_status_date')['adr'].mean()

# Plotting the ADR over time for both hotels
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel')

# Plot for Resort Hotel
plt.plot(Resort_hotel_avg.index, Resort_hotel_avg, label='Resort Hotel')

# Plot for City Hotel
plt.plot(city_hotel_avg.index, city_hotel_avg, label='city Hotel')

# Add labels and legend
plt.xlabel('Reservation Status Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.legend()

# Display the plot
plt.show()
```

```python
# Ensure 'reservation_status_date' is in datetime format
data['reservation_status_date'] = pd.to_datetime(data['reservation_status_date'])

# Create a 'month' column from the 'reservation_status_date' column
data['month'] = data['reservation_status_date'].dt.month

# Plot the count of reservations per month, categorized by 'is_canceled'
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=data)

# Show the plot
plt.show()
```

```python
# Filter for canceled reservations
cancelled_data = data[data['is_canceled'] == 1]

# Get the top 10 countries with canceled reservations
top_10_country = cancelled_data['country'].value_counts()[:10]

# Plotting the pie chart
plt.figure(figsize=(20, 8))
plt.title('Top 10 Countries with Canceled Reservations')

# Create the pie chart
plt.pie(top_10_country, labels=top_10_country.index, autopct='%.2f%%')

# Show the plot
plt.show()
```

```python
# Group the data by 'month' and sum the 'adr', but reset the index so 'month' is a column
adr_per_month = data[data['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()

# Plot
plt.figure(figsize=(15,8))
plt.title('ADR per Month for Canceled Reservations', fontsize=30)
sns.barplot(x='month', y='adr', data=adr_per_month)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Total ADR', fontsize=15)

# Show the plot
plt.show()
```

```python
# Group and calculate mean ADR for canceled reservations
cancelled_data_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean().reset_index()
cancelled_data_adr.sort_values('reservation_status_date', inplace=True)

# Group and calculate mean ADR for not canceled reservations
notcancelled_data = data[data['is_canceled'] == 0]
notcancelled_data_adr = notcancelled_data.groupby('reservation_status_date')[['adr']].mean().reset_index()
notcancelled_data_adr.sort_values('reservation_status_date', inplace=True)

# Plotting
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate', fontsize=30)

# Plot for not canceled reservations
plt.plot(notcancelled_data_adr['reservation_status_date'], notcancelled_data_adr['adr'], label='Not Canceled')

# Plot for canceled reservations
plt.plot(cancelled_data_adr['reservation_status_date'], cancelled_data_adr['adr'], label='Canceled')

# Adding labels and legend
plt.xlabel('Reservation Status Date', fontsize=15)
plt.ylabel('Average Daily Rate (ADR)', fontsize=15)
plt.legend()
plt.grid()  # Optional: Add grid for better readability

# Show the plot
plt.show()
```
